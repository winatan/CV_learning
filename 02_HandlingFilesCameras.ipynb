## Import Necessary Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

### Create a Gray Image

```python
img = np.zeros((3,3), dtype=np.uint8)
img
```

### Convert Image to Blue-Green-Red (BGR) format from Gray

```python
img = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
img
```

### Read in an Image

```python
img = cv2.imread('./Assets/Frieren.png')
print(img.shape)
print(img.size)
print(img.dtype)
cv2.imwrite('./Generated/CO2_testing.jpg',img)
```

```python
## Option to load an image in grascale at half its oritignal resoultion 
img = cv2.imread('./Assets/Frieren.png', cv2.IMREAD_REDUCED_GRAYSCALE_2)
cv2.imwrite('./Generated/CO2_testing.jpg',img)
```

```python
byteArray = bytearray(img)
grayImage = np.array(byteArray).reshape(img.shape[0], img.shape[1])
cv2.imwrite('./Generated/CO2_testing2.jpg', grayImage)
```

### Accessing Image Data with numpy.array

```python
img = cv2.imread('./Assets/Frieren.png', cv2.IMREAD_UNCHANGED)
img.itemset((0,0,1), 255) # sets the value at the specified channel index (B/G/R)
print(img.item(0,0,0)) # retrieve the value at pixel at (150,120) and the channel index 0 (Blue)

# Possible to also use python slicing to change a whole lot of pixel color
img[:100,:, 1] = 0
cv2.imwrite('./Generated/CO2_testing3.jpg', img)
plt.imshow(img)
```
