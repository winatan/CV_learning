## Import Necessary Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
from scipy import ndimage
```

## Basics of Low Pass & High Pass Filters (HPFs & LPFs)

```python
# High Pass Filters
kernel_3x3 = np.array([[-1,-1,-1],
                       [-1, 8,-1],
                       [-1,-1,-1]])
kernel_5x5 = np.array([[-1,-1,-1,-1,-1],
                       [-1, 1, 2, 1,-1],
                       [-1, 2, 4, 2,-1],
                       [-1, 1, 2, 1,-1],
                       [-1,-1,-1,-1,-1]])
```

```python
img = cv2.imread("./Assets/txtbk_images/statue_small.jpg",0) #load image in grey_Scale
# High Pass Filter
k3 = ndimage.convolve(img,kernel_3x3)
k5 = ndimage.convolve(img,kernel_5x5)
blurred = cv2.GaussianBlur(img,(17,17),0)
# Subtract the Low Pass Filter from Image
g_hpf = img - blurred
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,8))

axs[0,0].imshow(k3)
axs[0,0].set_title('3x3 Kernel')

axs[0,1].imshow(k5)
axs[0,1].set_title('5x5 Kernel')

axs[1,0].imshow(blurred)
axs[1,0].set_title('Blurred - (Low Pass Filter)')

axs[1,1].imshow(g_hpf)
axs[1,1].set_title('g_hpf (Complement Method)')
plt.tight_layout()
```
